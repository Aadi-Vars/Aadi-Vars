```python
import pandas as pd
```

```python
male_df=pd.read_csv("Male-Names.csv")
```

```python
male_df.info()
```

```python
male_df.describe()
```

```python
male_df.drop(columns=["race"],inplace=True)
```

```python
male_df["name"]=male_df["name"].astype("string")
```

```python
male_df["gender"]=male_df["gender"].astype("string")
```

```python
male_df=male_df.dropna()
```

```python
male_df.info()
```

```python
male_df=male_df.drop_duplicates()
```

```python
male_df.info()
```

```python
male_df
```

```python
for i in male_df.index.values:
    male_df.at[i,"name"]=male_df.at[i,"name"].split(" ")[0]
    if "@" in male_df.at[i,"name"]:
       male_df.at[i,"name"]=male_df.at[i,"name"].split("@")[0]
```

```python
for i in male_df.index.values:
    if len(male_df.at[i,"name"])<3:
       male_df=male_df.drop(i)
```

```python
male_df.to_csv("male_final.csv",index=None)
```
