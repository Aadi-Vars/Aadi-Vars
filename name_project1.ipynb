```python
import pandas as pd
```

```python
listy=[]
for i in range(65,91):
    listy.append(chr(i))
for i in range(97,123):
    listy.append(chr(i))
```

```python
male_df=pd.read_csv("Male-Names.csv")
```

```python
male_df.info()
```

```python
male_df.describe()
```

```python
male_df.drop(columns=["race"],inplace=True)
```

```python
male_df["name"]=male_df["name"].astype("string")
```

```python
male_df["gender"]=male_df["gender"].astype("string")
```

```python
male_df=male_df.dropna()
```

```python
male_df.info()
```

```python
male_df
```

```python
for i in male_df.index.values:
    value=male_df.at[i,"name"]
    male_df.at[i,"name"]=value.split(" ")[0].upper()
    value=value.split(" ")[0].upper()
    if "@" in value:
       male_df.at[i,"name"]=value.split("@")[0]
       value=value.split("@")[0]
    if len(value)<=3:
       male_df=male_df.drop(i)
       continue
    if not (value.strip(" ")[0] in listy):
      male_df=male_df.drop(i)
      continue
    if "-" in value:
       if len(value.split("-")[0])<=3:
           male_df.at[i,"name"]=value.split("-")[1]
           value=value.split("-")[1]
       else:
           male_df.at[i,"name"]=value.split("-")[0]
           value=value.split("-")[0]
```

```python
male_df=male_df.drop_duplicates()
```

```python
male_df.describe()
```

```python
male_df
```
